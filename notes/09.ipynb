```python
#numbers = [1, 2, 3, 4]
#mean(numbers)

def mean(numbers):
    return sum(numbers) / len(numbers)

mean([1, 2, 3, 4])
```

```python
def mean(numbers):
    return sum(numbers) / len(numbers)

def std(numbers):
    licznik = 0
    average = mean(numbers)
    for number in numbers:
        licznik += (number - average) ** 2
    mianownik = len(numbers)
    return (licznik / mianownik) ** 0.5

numbers = [5, 5, 5, 6, 4, 4, 7, 5]
std(numbers)
```

```python
def multiply(*args):
    # * - zbiera argumenty w krotkę
    prod = 1
    for number in args:
        prod *= number
    return prod

multiply(1, 2, 3, 4)
```

```python
def multiply(**kwargs):
    # ** - zbiera argumenty w słownik
    print(kwargs)

multiply(a=1, b=2, c=4)
```
