```python
def std(sample):
    mean = sum(sample) / len(sample)
    licznik = 0
    for number in sample:
        licznik += (number - mean)**2
    mianownik = len(sample)
    return (licznik / mianownik) ** 0.5
```

```python
std([1 ,2, 8, 11, 30])
```

```python
def gc_content(seq):
    '''Calculate G+C content.

    Args:
       seq (str) - input sequence

    Returns:
        value between 0 and 1.
    '''
    seq = seq.upper()
    g = seq.count('G')
    c = seq.count('C')
    return (g + c) / len(seq)
```
