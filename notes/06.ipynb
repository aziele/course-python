```python
dic = {'hi': 'cześć', 'bye': 'na razie'}

if 'one' in dic:
    print(dic['one'])
else:
    print('Sorry, no translation')
```

```python
print('one' in dic)
print('one' not in dic)
```

```python
seq = 'MGSTWPRSQTTAAACCCVWE'

counts = {}
for aminoacid in seq:
    if aminoacid not in counts:
        counts[aminoacid] = 1
    else:
        counts[aminoacid] += 1

print(counts)
```

## Iterowanie po słowniku

```python
for key in dic:
    print(key)
```

```python
d = {'Python': 1, 'C++': 2, 'JavaScript': 3}
for language in d:
    print(language, d[language])
```

```python
d = {'Python': 1, 'C++': 2, 'JavaScript': 3}
for language in sorted(d.keys()):
    print(language, d[language])
```

```python
d = {'Python': 1, 'C++': 2, 'JavaScript': 3}
for language in sorted(d):
    print(language, d[language])
```

```python
d.items()
```

```python
for language, language_number in d.items():
    print(language, language_number)
```
