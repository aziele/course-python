```python
class SeqRecord:

    def __init__(self, id, seq, description=None):
        self.id = id
        self.seq = seq
        self.description = description

    def fasta(self):
        if self.description:
            desc = self.description
        else:
            desc = ''
        return f'>{self.id} {desc}\n{self.seq}'

    def counts(self):
        d = {}
        for char in self.seq:
            if char not in d:
                d[char] = 1
            else:
                d[char] += 1
        return d

    def __str__(self):
        return self.id

    def __len__(self):
        return len(self.seq)

    def __iter__(self):
        return iter(self.seq)


my_record = SeqRecord('geneX', 'ATGCTAGTA', 'protein-coding gene')

len(my_record)

```

```python
class DNARecord(SeqRecord):

    def transcribe(self):
        return self.seq.replace('T', 'U')
```

```python
record = DNARecord(id='geneX', seq='ATGATGA')
print(record.counts())
```

```python
print(record.transcribe())
```
