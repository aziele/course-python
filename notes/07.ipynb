## Wczytywanie danych z pliku

## 1 sposób

```python
f = open('infile.txt')
for line in f:
    print(line)
f.close()
```

```python
f = open('infile.txt')
for line in f:
    print(line.upper())
f.close()
```

## 2 Sposób

```python
f = open('infile.txt')
lines = f.readlines()
f.close()
print(lines)
```

```python
f = open('infile.txt')
lines = []
for line in f:
    lines.append(line)
print(lines)
```

## 3 sposób

```python
f = open('infile.txt')

line = f.readline()
print(line)

line = f.readline()
print(line)

f.close()
```

### 4 sposób

```python
f = open('infile.txt')

content = f.read()
f.close()

print(content.upper())
```

```python
f = open('infile.txt')

content = f.read(100)
print(content)

print()
content = f.read(100)
print(content)

print()
content = f.read(100)
print(content)

f.close()
```

```python
# 1: pętla for: wiersz po wierszu
# 2: readlines(): wszystkei wiersze do listy
# 3: readline(): wczytuje jako string pojedynczy wiersz
# 4: read(): wczytuje zawartość całego pliku jako string

```

# Zapisywanie do pliku

```python
f = open('outfile.txt', 'w')

f.write('Test')

f.close()
```

```python
!cat outfile.txt
```

```python
f = open('outfile.txt', 'w')

f.write('Line1\n')
f.write('Line2\n')

f.close()
```

```python
!cat outfile.txt
```

```python
lines = ['Line1\n', 'Line2\n', 'Line3\n']

f = open('outfile.txt', 'w')
f.writelines(lines)
f.close()
```

```python
!cat outfile.txt
```

```python
lines = ['Line1\n', 'Line2\n', 'Line3\n']

f = open('outfile.txt', 'w')
for line in lines:
    f.write(line)
f.close()
```

```python
!cat outfile.txt
```

```python
f = open('infile.txt')
output_file = open('results.txt', 'w')
for line in f:
    output_file.write(f'{len(line)}\n')
f.close()
output_file.close()
```

```python
!cat results.txt
```

```python
max_word_size = 0

f = open('words_alpha.txt')
for line in f:
    word_size = len(line.rstrip())
    if word_size > max_word_size:
        max_word_size = word_size
        max_word = line
f.close()

print(max_word)
print(max_word_size)

```
