# Nowe generatory,których nie było

```python
name = 'World'

i = 0
for char in name:
    print(char, i)
    i += 1
```

```python
name = 'World'
for i, char in enumerate(name):
    print(char, i)
```

```python
list(enumerate(name))
```

```python
name = 'World'
for i, char in enumerate(name, start=1):
    print(char, i)
```

```python
countries = ['Poland', 'Germany', 'USA']
cities = ['Poznan', 'Berlin', 'New York']

for i in range(len(cities)):
    city = cities[i]
    country = countries[i]
    print(country, city)
```

```python
countries = ['Poland', 'Germany', 'USA']
cities = ['Poznan', 'Berlin', 'New York']

list(zip(countries, cities))
```

```python
for country, city in zip(countries, cities):
    print(country, city)
```

```python
seq1 = 'AAAAA'
seq2 = 'BBB'
seq3 = 'CC'

for l1, l2, l3 in zip(seq1, seq2, seq3):
    print(l1, l2, l3)
```

```python
seq1 = 'ATGCTGGTAG'
seq2 = 'ATGCAGGTTG'

distance = 0
for nt1, nt2 in zip(seq1, seq2):
    if nt1 != nt2:
        distance += 1

print(distance)
```

```python
len([1 for nt1, nt2 in zip(seq1, seq2) if nt1 != nt2])
```

# Tworzenie własnych generatorów

```python
def sqrange(n):
    numbers = []
    for number in range(n):
        numbers.append(number ** 2)
    return numbers
```

```python
sqrange(10)
```

```python
def sqrange(n):
    for number in range(n):
        yield number ** 2
```

```python
sqrange(10)
```

```python
for number in sqrange(10):
    print(number)
```

```python
list(sqrange(10))
```

```python
generator = sqrange(5)
for number in generator:
    print(number)

for number in generator:
    print(number)
```

# Funkcja anonimowa lamda

```python
def power1(number):
    return number ** 2

power1(2)
```

```python
power2 = lambda number: number ** 2

power2(2)
```

## Sortowanie i lambda

```python
lst = ['elephant', 'boy', 'apple']
lst.sort()
print(lst)
```

```python
lst = ['elephant', 'boy', 'apple']
lst.sort(key=len)
print(lst)
```

```python
def count_e(word):
    return word.count('e')

lst = ['elephant', 'boy', 'apple']
lst.sort(key=count_e)
print(lst)
```

```python
lst = ['elephant', 'boy', 'apple']
lst.sort(key=lambda word: word.count('e'))
print(lst)
```

```python
lst = [
    ('NOT1', 340),
    ('GW182', 10),
    ('NOT1', 2),
    ('FOXP2', 55),
    ('MET23', 55),
    ('BRCA2', 55)
]

lst.sort()
print(lst)
```

```python
lst.sort(key=lambda x: x[1])
print(lst)
```

```python
lst.sort(key=lambda x: x[1], reverse=True)
print(lst)
```

```python
lst = [
    ('NOT1', 340),
    ('GW182', 10),
    ('NOT1', 2),
    ('FOXP2', 55),
    ('MET23', 55),
    ('BRCA2', 55)
]

new_lst = sorted(lst, key=lambda x: x[1], reverse=True)
```

```python
d = {
    'A': 240,
    'T': 10,
    'C': 1,
    'G': 5,
}

for nt, count in sorted(d.items(), key=lambda tup: tup[1], reverse=True):
    print(nt, count)
```
