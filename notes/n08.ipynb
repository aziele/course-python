# 1. Tworzenie własnych modułów

```python
# dna.py
version = '1.0'

def complement(dna):
    d = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
    return "".join([d.get(nt, 'N') for nt in dna])

def reverse_complement(dna):
    return complement(dna)[::-1]

if __name__ == '__main__':
    user_seq = input('Enter a DNA sequence: ')
    revcomp_seq = reverse_complement(user_seq)
    print(f'Reverse complement: {revcomp_seq}')
```

## 2.3. Moduł itertools

```python
import itertools

nucleotides = 'ABCD'

for tup in itertools.permutations(nucleotides):
    print(tup)
```

```python
for tup in itertools.permutations(nucleotides, r=3):
    print(tup)
```

```python
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]

for number1 in numbers1:
    for number2 in numbers2:
        print(number1, number2)
```

```python
for number1, number2 in itertools.product(numbers1, numbers2):
    print(number1, number2)
```

```python
people = 'ABCDE'
# Ile będzie podań ręki między 5 osobami
for tup in itertools.combinations(people, 2):
    print(tup)
```
