```python
def is_dna(seq):
    for nucleotide in seq:
        if nucleotide not in {'A', 'T', 'G', 'C'}:
            return False
    return True
```

```python
def word_count(seq, word_size=1, normalize=False):
    """Count words of given size in DNA sequence.

    Args:
        seq (str): DNA sequence
        word_size (int): Word size (k-mer size)

    Returns:
        Dictionary with word/k-mer counts
    """
    counts = {}
    for i in range(len(seq) - word_size + 1):
        word = seq[i:i + word_size]
        if word not in counts:
            counts[word] = 1
        else:
            counts[word] += 1

    if normalize:
        total_counts = sum(counts.values())
        for word in counts:
            counts[word] = counts[word] / total_counts * 100

    return counts

print(word_count('ATGAGATGACGGTAG', 3, True))
```
