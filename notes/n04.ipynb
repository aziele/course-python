Działania matematyczne

```python
dna = set('ATGC')
rna = set('AUGC')

print(dna)
print(rna)
```

```python
dna | rna
```

```python
dna - rna
```

```python
dna & rna
```

```python
dna ^ rna
```

```python
intersection = dna & rna
print(intersection)
```

```python
nucleotides = dna | rna
print(nucleotides)
```

```python
dna < nucleotides
```

```python
nucleotides > dna
```

```python
dna.union(rna)    # dna | rna
```

```python
dna.difference(rna)   # dna - rna
```

```python
dna.intersection(rna)   # dna & rna
```

```python
dna.symmetric_difference(rna)  # dna ^ rna
```

```python
protein = 'MGCDEFGLIKMNPRSTQWVYFGSTKSTWQAG'

counts = {}
for aminoacid in protein:
    if aminoacid not in counts:
        counts[aminoacid] = 1
    else:
        counts[aminoacid] += 1

print(counts)
```

```python
protein = 'MGCDEFGLIKMNPRSTQWVYFGSTKSTWQAG'

counts = {}
for aminoacid in protein:
    if aminoacid not in counts:
        counts[aminoacid] = 0
    counts[aminoacid] += 1


print(counts)
```

```python
protein = 'MGCDEFGLIKMNPRSTQWVYFGSTKSTWQAG'

counts = {}
for aminoacid in protein:
    counts[aminoacid] = counts.get(aminoacid, 0) + 1


print(counts)
```
