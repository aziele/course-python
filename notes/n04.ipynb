Zbiór jest nieuporządkowaną kolekcją niemutawalnych elementów, które się nie powtarzają.

```python
lst = ['geneA', 'geneB', 'geneB', 'geneC', 'geneA', 'geneA']
print(list(set(lst)))
```

```python
lst = ['geneA', 'geneB', 'geneB', 'geneC', 'geneA', 'geneA']
new_lst = []
for gene in lst:
    if gene not in new_lst:
        new_lst.append(gene)
print(new_lst)
```

```python
protein1 = 'MGSWPSTT'
protein2 = 'MGAPVHIK'

print(set(protein2) - set(protein1))
```

```python
lotto = set()
while len(lotto) < 6:
    number = random.randint(1, 50)
    lotto.add(number)
print(lotto)
```

```python
lotto = []
while len(lotto) < 6:
    number = random.randint(1, 50)
    if number not in lotto:
        lotto.append(number)
print(lotto)
```

```python
sequence = 'MGSTGGGWWWPPPSSSTPAAGGLLIIITW'
counts = {}

for aminoacid in sequence:
    if aminoacid not in counts:
        counts[aminoacid] = 1
    else:
        counts[aminoacid] += 1

print(counts)
```

```python
sequence = 'MGSTGGGWWWPPPSSSTPAAGGLLIIITW'
counts = {}

for aminoacid in sequence:
    if aminoacid not in counts:
        counts[aminoacid] = 0
    counts[aminoacid] += 1

print(counts)
```

```python
sequence = 'MGSTGGGWWWPPPSSSTPAAGGLLIIITW'
counts = {}

for aminoacid in sequence:
    counts[aminoacid] = counts.get(aminoacid, 0) + 1

print(counts)
```
