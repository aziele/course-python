## Działania matematyczne

```python
dna = set('ATGC')
rna = set('AUGC')

print(dna)
print(rna)
```

```python
print(dna - rna)
```

```python
print(dna | rna)
```

```python
print(dna & rna)
```

```python
print(dna ^ rna)
```

```python
nucleotides = dna | rna
print(nucleotides)
```

```python
print(nucleotides > dna)
```

```python
'A' in dna
```

```python
s = 'ATGCTAGTAG'
print(set(s))
```

```python
list(s)
```

```python
print(dna.difference(rna))
```

```python
print(dna.union(rna))
```

```python
print(dna.intersection(rna))     # dna & rna
```

```python
print(dna.symmetric_difference(rna))  # dna ^ rna
```

```python
nucleotides.issuperset(dna)   # nucleotides > dna
```

```python
nucleotides.issubset(dna)    # nucleotides < dna
```

```python
import random

print(random.randint(1, 49))
```

```python
lotto = set()
while len(lotto) < 6:
    number = random.randint(1, 49)
    lotto.add(number)

print(lotto)
```

```python
lotto = []
while len(lotto) < 6:
    number = random.randint(1, 49)
    if number not in lotto:
        lotto.append(number)

print(lotto)
```

# Obsługa wyjątków

```python
temp = input('Enter a temperature:')
try:
    temp = float(temp)
except ValueError:
    print('Sorry I could not convert it to number')
except Exception:
    print('Something went wrong')
else:
    f = 32 + 9/5 * temp
    print(f'Temperature in F: {f}')
finally:
    print('All done')
```
