```python
print('Type `quit` to exit.')
numbers_count = 0
numbers_sum = 0
while True:
    user_input = input('Enter a number: ')
    if user_input.lower() == 'quit':
        break
    numbers_count += 1
    numbers_sum += int(user_input)

if numbers_count:
    print(f'Numbers : {numbers_count}')
    print(f'Sum     : {numbers_sum}')
    print(f'Mean    : {numbers_sum / numbers_count:.1f}')
```

```python
seq1 = 'ACGTAGATAGTAGATA'
seq2 = 'ACGTAGCTAGTAGAAA'

distance = 0
for position in range(0, len(seq1)):
    letter1 = seq1[position]
    letter2 = seq2[position]
    if letter1 != letter2:
        distance += 1

print(f'Hamming distance: {distance}')
```
