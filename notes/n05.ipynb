```python
f = open('words_alpha.txt')
total_words = 0
for line in f:
    if 'e' in line:
        total_words += 1
f.close()
print(f'Words: {total_words:,}')
```

```python
f = open('words_alpha.txt')
max_length = 0
max_word = ''
for line in f:
    word = line.rstrip()
    word_length = len(word)
    if word_length > max_length:
        max_length = word_length
        max_word = word
f.close()
print(f'Max word: {max_word}')
print(f'Max word length: {max_length}')
```

```python
f = open('words_alpha.txt')
counts = {}
for line in f:
    word = line.rstrip()
    word_length = len(word)
    if word_length not in counts:
        counts[word_length] = 0
    counts[word_length] += 1
f.close()

for word_length in sorted(counts):
    print(f'{word_length} {counts[word_length]}')
```

```python
f = open('words_alpha.txt')
counts = {}
for line in f:
    word = line.rstrip()
    word_length = len(word)
    if word_length not in counts:
        counts[word_length] = []
    counts[word_length].append(word)
f.close()

print(counts[27])
```

```python
f = open('Escherichia_coli.tsv')
out_file = open('results.tsv', 'w')
out_file.write('gene_name\tgene_length\n')
f.readline()
for line in f:
    columns = line.split()
    gene_name = columns[0]
    start = int(columns[1])
    end = int(columns[2])
    length = end - start + 1
    out_file.write(f'{gene_name} {length}\n')
f.close()
out_file.close()
```

# 2. Obsługa wyjątków

```python
user_input = input('Enter a temperature in Celsius: ')
c = int(user_input)
f = 32 + 9 / 5 * c
print(f'Fahrenheit: {f:.2f}')
```

```python
user_input = input('Enter a temperature in Celsius: ')
try:
    c = int(user_input)
    f = 32 + 9 / 5 * c
    print(f'Fahrenheit: {f:.2f}')
except Exception:
    print('Sorry, something went wrong')
```

```python
user_input = input('Enter a temperature in Celsius: ')
try:
    c = int(user_input)
except ValueError:
    print(f'Sorry, could not convert {user_input} to integer.')
except Exception:
    print('Sorry, something went wrong')
else:
    f = 32 + 9 / 5 * c
    print(f'Fahrenheit: {f:.2f}')
finally:
    print('I am always executed.')
```
