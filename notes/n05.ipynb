## 1.3. Kilka przykładów pracy z plikami

```python
f = open('words_alpha.txt')
i = 0
for line in f:
    i += 1
f.close()
print(f'Words: {i:,}')
```

```python
f = open('words_alpha.txt')
i = 0
for line in f:
    if 'e' in line:
        i += 1
f.close()
print(f'Words with e: {i:,}')
```

```python
f = open('words_alpha.txt')
max_length = 0
max_word = ''
for word in f:
    word = word.rstrip()
    word_length = len(word)
    if word_length > max_length:
        max_length = word_length
        max_word = word
f.close()

print(f'Max word: {max_word}')
print(f'Max len : {max_length}')
```

```python
f = open('words_alpha.txt')
# Budowanie słownika zliczeń
counts = {}
for line in f:
    word = line.rstrip()
    word_size = len(word)
    if word_size not in counts:
        counts[word_size] = 1
    else:
        counts[word_size] += 1
f.close()

# Wyświetlenie zawartości słownka
for word_size in sorted(counts):
    print(f'{word_size} {counts[word_size]}')
```
