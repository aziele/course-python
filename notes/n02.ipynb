Instrukcja warunkowe

```python
language = input('What is your favorite programming language? ')
if language.lower() == 'python':
    print('Condition was True')
    print('I like Python too')
print('Thanks!')
```

```python
language = input('What is your favorite programming language? ')
if language.lower() == 'python':
  print('I like Python too')
else:
  print('No match')
```

```python
number = int(input('Enter a number: '))
if number % 2 == 0:
  print(f'Number {number} is even.')
else:
  print(f'Number {number} is odd.')
```

```python
age = int(input('How old are you? '))
if age >= 18:
  print('You are an adult.')
elif age == 100:
  print('Wow, you are a century old!')
elif age < 0:
  print('You have not been born yet.')
else:
  if age < 3:
    print('You are a baby')
  print('You are a child!')
```

**Pętla while**

```python
x = 1
if x < 10:
  print(x)
```

```python
x = 1
while x < 10:
  print(x)
  x += 1
print('Done')
```

```python
x = 1
while x < 10:
  print(x)
```

```python
while True:
  option = input('Choose option a or b: ')
  if option == 'a' or option == 'b':
    break
    print('test')

print(f'Thanks for choosing: {option} ')
```

```python
x = 0
while x < 10:
  x += 1
  if x % 2 == 0:
    continue
  print(x)
```

```python
secret_number = 7

guess_count = 0
while guess_count < 3:
  user_number = int(input('Guess a number: '))
  guess_count += 1
  if user_number == secret_number:
    print('You won!')
    break
else:
  print('You are failed')
```
