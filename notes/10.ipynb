# Zasięgi nazw

```python
# 1. Lokalny
# 2. Globalny
# 3. Built-in (zmienne wbudowane w Pythonie, np. len)
```

```python
x = 10            # Zmienna globalna

def f():
    y = 5         # Zmienna lokaln
    print(x)
    print(y)

f()
```

```python
x = 10            # Zmienna globalna

def f():
    y = 5         # Zmienna lokaln
    print(x)
    print(y)

f()
print(y)
```

```python
x = 10            # Zmienna globalna

def f():
    x = 5         # Zmienna lokaln
    print(x)

f()
print(x)
```

# Listy składane (list comprehension)

```python
minutes = [2, 3, 1, 4]

seconds = []
for minute in minutes:
    seconds.append(minute * 60)

print(seconds)
```

```python
seconds = [minute * 60 for minute in minutes]
print(seconds)
```

```python
genes = ['TNRC6A', 'BRCA2', 'WAK1', 'RRM2']

human_genes = [f'Hs{gene}' for gene in genes]

print(human_genes)
```

```python
human_genes = []
for gene in genes:
    human_genes.append(f'Hs{gene}')
print(human_genes)
```

```python
lst = [i ** 2 for i in range(10)]
print(lst)
```

```python
lst = [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

lst_string = [str(number) for number in lst]
print(lst_string)
```

# Filtrowanie

```python
lst = [1, 2, 85, 84, 5, 6, 10, 11, 32, 64]

even = [number for number in lst if number % 2 == 0]
print(even)
```

```python
s1 = 'abc'
s2 = 'def'

lst = [char1 + char2 for char1 in s1 for char2 in s2 if char1 != char2] # Nieczytelne
print(lst)
```

```python
genes = ['TNRC6A', 'BRCA2', 'WAK1', 'RRM2']

d = {gene: gene.lower() for gene in genes}
print(d)
```

```python
myset = {gene for gene in genes}
print(myset)
```
