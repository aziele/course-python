## 1.2. Przydatne generatory

### 1.2.1 enumerate

```python
seq = 'ATGTCGTA'
position = 0
for nucleotide in seq:
    print(nucleotide, position)
    position += 1
```

```python
seq = 'ATGTCGTA'
for position, nucleotide in enumerate(seq):
    print(nucleotide, position)
```

```python
print(list(enumerate(seq)))
```

```python
seq = 'ATGTCGTA'
for position, nucleotide in enumerate(seq, start=1):
    print(nucleotide, position)
```

```python
lst = ['Gene1', 'Gene2', 'Gene3']
for i, gene_name in enumerate(lst):
    print(i, gene_name)
```

# 1.2.2. zip

```python
seq1 = 'ATCGTAG'
seq2 = 'ATCCTTG'

distance = 0
for i in range(len(seq1)):
    nt1 = seq1[i]
    nt2 = seq2[i]
    if nt1 != nt2:
        distance += 1
print(distance)
```

```python
list(zip(seq1, seq2))
```

```python
seq1 = 'ATCGTAG'
seq2 = 'ATCCTTG'

distance = 0
for nt1, nt2 in zip(seq1, seq2):
    if nt1 != nt2:
        distance += 1
print(distance)
```

```python
seq1 = 'ATCGTAG'
seq2 = 'ATCCTTG'
seq3 = 'ATGCTAG'

for nt1, nt2, nt3 in zip(seq1, seq2, seq3):
    print(nt1, nt2, nt3)
```

```python
seq1 = 'ATCGTAG'
seq2 = 'ATCCTTG'
seq3 = 'ATG'

for nt1, nt2, nt3 in zip(seq1, seq2, seq3):
    print(nt1, nt2, nt3)
```

```python
countries = ['Poland', 'Germany', 'USA']
cities = ['Warsaw', 'Berlin', 'Washington']

for country, city in zip(countries, cities):
    print(country, city)
```

## 2. Tworzenie własnych generatorów

```python
def sqrange(n):
    lst = []
    for number in range(n):
        lst.append(number ** 2)
    return lst

print(sqrange(10))
```

```python
def sqrange(n):
    for number in range(n):
        yield number ** 2

print(sqrange(10))
```

```python
for i in sqrange(10):
    print(i)
```

```python
list(sqrange(10))
```

```python
r = sqrange(5)
for number in r:
    print(number)
```

```python
for number in r:
    print(number)
```

```python
r = sqrange(5)
for number in r:
    print(number)

r = sqrange(5)
for number in r:
    print(number)
```

```python
r = list(sqrange(5))
for number in r:
    print(number)

for number in r:
    print(number)
```

# 2. Składanie list (list comprehension)

```python
minutes = [2, 3, 5, 10, 11]
```

```python
seconds = []
for m in minutes:
    seconds.append(m * 60)
print(seconds)
```

```python
seconds = [m * 60 for m in minutes]
print(seconds)
```

```python
numbers = list(range(10))
print(numbers)
```

```python
lst = [number ** 2 for number in numbers]
print(lst)
```

```python
row = ['1', '2', '100', '1000', '200']
row = [int(value) for value in row]
print(row)
```

```python
genes = ['Fox2P', 'GW182', 'BRCA2', 'NOT1']
human_genes = [f'Hs{gene}' for gene in genes]
print(human_genes)
```

```python
numbers = [1, 2, 0, 3, 7, 100, 12, 11, 13]
even_numbers = [number for number in numbers if number % 2 == 0]
print(even_numbers)
```

```python
genes
```

```python
{gene.upper(): gene for gene in genes}
```

```python
{gene.upper() for gene in genes}
```

```python
names = ['Bruce', 'Clark', 'Peter']
heroes = ['Batman', 'Superman', 'Spiderman']

d = {name: hero for name, hero in zip(names, heroes) if name != 'Bruce'}
print(d)
```

```python
lst = [i ** 2 for i in range(10)]
gen = (i ** 2 for i in range(10))

print(gen)
```

# Funkcja anonimowa lambda

```python
def power1(number):
    return number ** 2

power2 = lambda number: number ** 2
```

```python
print(power1(10))
print(power2(10))
```

```python
lst = ['elephant', 'boy', 'apple']
lst.sort()
print(lst)
```

```python
lst.sort(key=len)
print(lst)
```

```python
lst.sort(key=len, reverse=True)
print(lst)
```

```python
lst = ['elephant', 'boy', 'apple']
```

```python
def count_e(word):
    return word.count('e')

lst.sort(key=count_e)
print(lst)
```

```python
lst.sort(key=lambda word: word.count('e'))
print(lst)
```

```python
lst = [
   ('NOT1', 340),
   ('GW182', 10),
   ('NOT1', 2),
   ('FOXP2', 55),
   ('MET23', 55),
   ('BRCA2', 55)
]

lst.sort()
print(lst)
```

```python
lst.sort(key=lambda x: x[1])
print(lst)
```

```python
lst.sort(key=lambda x: (x[1], x[0]))
print(lst)
```

```python
lst.sort(key=lambda x: (-x[1], x[0]))
print(lst)
```

```python
d = {'A': 30, 'C': 50, 'G': 40, 'T': 20}
for key, value in sorted(d.items(), key=lambda x: x[1], reverse=True):
    print(key, value)
```
