### zip

```python
seq1 = 'ATGCTGA'
seq2 = 'ATGCAGT'

distance = 0
for i in range(len(seq1)):
    letter1 = seq1[i]
    letter2 = seq2[i]
    if letter1 != letter2:
        distance += 1

print(distance)
```

```python
list(zip(seq1, seq2))
```

```python
seq1 = 'ATGCTGA'
seq2 = 'ATGCAGT'

distance = 0
for letter1, letter2 in zip(seq1, seq2):
    if letter1 != letter2:
        distance += 1

print(distance)
```

```python
seq1 = 'ATGCTGA'
seq2 = 'ATGCAGT'
seq3 = 'ATGCTAG'

for l1, l2, l3 in zip(seq1, seq2, seq3):
    print(l1, l2, l3)
```

```python
seq1 = 'ATGCTGA'
seq2 = 'ATGCAGT'
seq3 = 'ATG'

for l1, l2, l3 in zip(seq1, seq2, seq3):
    print(l1, l2, l3)
```

# 2. Tworzenie własnych generatorów

```python
def sqrange(n):
    lst = []
    for number in range(n):
        lst.append(number ** 2)
    return lst

print(sqrange(10))
```

```python
def sqrange(n):
    for number in range(n):
        yield number ** 2

print(sqrange(10))
```

```python
for i in sqrange(5):
    print(i)
```

```python
results = list(sqrange(5))
print(results)
```

```python
generator = sqrange(5)

for number in generator:
    print(number)

print('------')
for number in generator:
    print(number)
```

```python
import random

def random_codons(n):
    nucleotides = ['A', 'T', 'G', 'C']
    for i in range(n):
        nt1 = random.choice(nucleotides)
        nt2 = random.choice(nucleotides)
        nt3 = random.choice(nucleotides)
        yield f'{nt1}{nt2}{nt3}'

```

```python
for codon in random_codons(10):
    print(codon)
```

# 2. Listy składane (list comprehension)

```python
minutes = [1, 4, 5, 10, 100]
seconds = []
for minute in minutes:
    seconds.append(minute * 60)
print(seconds)
```

```python
seconds = [minute * 60 for minute in minutes]
print(seconds)
```

```python
lst = [i ** 2 for i in range(10)]
print(lst)
```

```python
genes = ['GW182', 'BRCA2', 'FOX2P', 'NOT1']
genes = [f'Hs{gene}' for gene in genes]
print(genes)
```

```python
row = ['2367', '69320', '7828732', '8493']
row = [int(value) for value in row]
print(row)
```

```python
numbers = [1, 3, 4, 6, 17, 13, 12, 15, 80]
numbers = [number for number in numbers if number % 2 == 0]
print(numbers)
```

# 3. Funkcja anonimowa lambda

```python
def power1(number):
    return number ** 2

power2 = lambda number: number ** 2

print(power1(10))
print(power2(10))
```

```python
lst = ['elephant', 'boy', 'apple']
lst.sort()
print(lst)
```

```python
lst.sort(key=len)
print(lst)
```

```python
lst.sort(key=len, reverse=True)
print(lst)
```

```python
def count_e(word):
    return word.count('e')

lst.sort(key=count_e)
print(lst)
```

```python
lst.sort(key=lambda word: word.count('e'))
print(lst)
```

```python
lst = [
    ('GW182', 340),
    ('NOT1', 2),
    ('FOX2P', 55),
    ('BRCA2', 55),
    ('NOT3', 1),
    ('MET23', 55)
]
lst.sort()
print(lst)
```

```python
lst.sort(key=lambda tup: tup[1])
print(lst)
```

```python
lst.sort(key=lambda tup: -tup[1])
print(lst)
```
