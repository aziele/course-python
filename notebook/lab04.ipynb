## Działania matematyczne

```python
dna = set('ATGC')
rna = set('AUGC')
```

```python
dna | rna  # suma zbiorów
```

```python
dna - rna  # różnica (nukleotydy charakterystyczne dla dna)
```

```python
rna - dna # różnica (nukleotydy charakterystyczna dla rna)
```

```python
dna ^ rna # różnica symetryczna
```

```python
dna & rna  # część wspólna
```

## Operatory porównania

```python
nucleotides = dna | rna
nucleotides
```

```python
nucleotides > dna
```

```python
nucleotides < dna
```

```python
'A' in dna
```

```python
'A' not in dna
```

```python
dna.union(rna)  # Suma (unia): dna | rna
```

```python
dna.difference(rna) # dna - rna
```

```python
dna.intersection(rna) # dna & rna
```

```python
dna.symmetric_difference(rna) # dna ^ rna
```

```python
nucleotides.issuperset(dna) # nucleotides > dna
```

```python
dna.issubset(nucleotides) # dna < nucleotides
```

```python
text = input('Enter text: ')
counts = {}
for letter in text:
    if letter not in counts:
        counts[letter] = 1
    else:
        counts[letter] += 1
print(counts)
```

```python
text = input('Enter text: ')
ALLOWED = 'abcdefghiklmnoprstuwxyz'
counts = {}
for letter in text:
    if letter in ALLOWED:
        if letter not in counts:
            counts[letter] = 1
        else:
            counts[letter] += 1
print(counts)
```

```python
text = input('Enter text: ')
counts = {}
for letter in text:
    if letter not in counts:
        counts[letter] = 0
    counts[letter] += 1
print(counts)
```
