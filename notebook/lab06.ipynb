```python
def gc_content(seq):
    seq = seq.upper()
    g = seq.count('G')
    c = seq.count('C')
    return (g + c) / len(seq)
```

```python
print(gc_content('ATGCTGATAGATGAGA'))
```

```python
print(gc_content('ATGTGAGAGATGCGTAGA'))
```

```python
lst = ['ATCGTA', 'ATGCTAGA', 'TGATAGTA']
for seq in lst:
    print(gc_content(seq))
```

```python
gc = gc_content('ATGATAGATGA')
```

```python
print(gc)
```

```python
def multiply(a, b, *args):
    product = a * b
    for number in args:
        product *= number
    return product
    
multiply(1, 3, 5, 2, 10, 99)
```

```python
lst = [1, 3, 5, 2, 10, 99]
multiply(*lst)
multiply(1, 3, 5, 2, 10, 99)
```

```python
def word_count(seq, word_size=1):
    '''Counts words of a given length in DNA sequence.
    
    Args:
       seq (str):  DNA sequence
       word_size (int): Word size / k-mer length
    
    Returns:
        Dictionary mapping words to their counts
        
    Example:
        word_count('ATCG', word_size=2)
        {'AT': 1, 'TC': 1, 'CG': 1}
    '''
    d = {}
    for i in range(len(seq) - word_size + 1):
        word = seq[i:i+word_size]
        if word not in d:
            d[word] = 0
        d[word] += 1
    return d
```

```python
word_count('ATGCTAGTAGATAG', 2)
```
