```python
suma = 0
counter = 0
while True:
    user_input = input('Enter a number: ')
    if user_input == 'quit':
        break
    number = int(user_input)
    suma += number
    counter += 1
    
print(f'Suma: {suma}')
print(f'Mean: {suma / counter}')
    
```

```python
seq1 = 'ATCGTAGATGATTT'
seq2 = 'ATGGTAGATGTTTA'
distance = 0
for i in range(len(seq1)):
    nucleotide1 = seq1[i]
    nucleotide2 = seq2[i]
    if nucleotide1 != nucleotide2:
        distance += 1
print(f'Hamming distance: {distance}')

```
