## enumarate

```python
seq = 'ATCGTAGATG'
position = 0
for nucleotide in seq:
    print(nucleotide, position)
    position += 1
```

## zip

```python
countries = ['Poland', 'Germany', 'USA']
cities = ['Poznan', 'Berlin', 'New York']
for i in range(len(cities)):
    city = cities[i]
    country = countries[i]
    print(country, city)
```

```python
list(zip(countries, cities))
```

```python
for country, city in zip(countries, cities):
    print(country, city)
```

```python
seq1 = 'ATCGTAG'
seq2 = 'GTCGATG'
seq3 = 'TAGAGTA'
for char1, char2, char3 in zip(seq1, seq2, seq3):
    print(char1, char2, char3)
```

```python
seq1 = 'ATCGTAG'
seq2 = 'GTCGATG'
seq3 = 'TA G'
for char1, char2, char3 in zip(seq1, seq2, seq3):
    print(char1, char2, char3)
```

### Dystans Hamminga

```python
seq1 = 'ATGCTAG'
seq2 = 'ATCCTTG'
distance = 0
for nt1, nt2 in zip(seq1, seq2):
    if nt1 != nt2:
        distance += 1
print(distance)
```

# Tworzenie generatorów

```python
def sqrange(n):
    lst = []
    for number in range(n):
        lst.append(number ** 2)
    return lst
    
```

```python
sqrange(5)
```

```python
def sqrange(n):
    for number in range(n):
        yield number ** 2
```

```python
sqrange(5)
```

```python
list(sqrange(5))
```

```python
for number in sqrange(5):
    print(number)
```

```python
generator = sqrange(5)
for number in generator:
    print(number)
for number in generator:
    print(number)
```

```python
import random
random.choice(['A', 'T', 'C', 'G'])
```

```python
def random_codons(n):
    nucleotides = list('ATGC')
    for i in range(n):
        nt1 = random.choice(nucleotides)
        nt2 = random.choice(nucleotides)
        nt3 = random.choice(nucleotides)
        yield f'{nt1}{nt2}{nt3}'
```

```python
random_codons(5)
```

```python
list(random_codons(5))
```

```python
for codon in random_codons(5):
    print(codon)
```

# Funkcje anonimowe (lambda)

```python
lst = ['elephant', 'boy', 'apple']
lst.sort()
lst
```

```python
lst.sort(reverse=True)
lst
```

```python
lst.sort(key=len)
print(lst)
```

```python
def count_e(word):
    return word.count('e')
```

```python
lst.sort(key=count_e)
print(lst)
```

```python
lst.sort(key=lambda word: word.count('e'))
lst
```

```python
count_e = lambda word: word.count('e')
```

```python
count_e('apple')
```
