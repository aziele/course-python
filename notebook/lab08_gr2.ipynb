# Moduł itertools

```python
import itertools
```

```python
people = 'ABCD'
list(itertools.combinations(people, 2))
```

```python
list(itertools.combinations(people, 3))
```

```python
for pair in itertools.combinations(people, 2):
    print(pair)
```

```python
numbers1 = [1, 4, 65]
numbers2 = [2, 1, 10]
numbers3 = [8, 9]
for n1 in numbers1:
    for n2 in numbers2:
        for n3 in numbers3:
            print(n1, n2, n3)
```

```python
list(itertools.product(numbers1, numbers2, numbers2))
```

```python
for n1, n2, n3 in itertools.product(numbers1, numbers2, numbers3):
    print(n1, n2, n3)
```

```python
people = 'ABCD'

for tup in itertools.permutations(people):
    print(tup)
```
