## Ćwiczenia

```python
line_count = 0
file = open('words_alpha.txt')
for line in file:
    line_count += 1
file.close()
print(line_count)
    
```

```python
file = open('words_alpha.txt')
lines = file.readlines()
print(len(lines))
```

```python
words = 0
file = open('words_alpha.txt')
for line in file:
    if 'e' in line:
        words += 1
file.close()
print(words)
```

```python
longest_word = ''
file = open('words_alpha.txt')
for line in file:
    if len(line) > len(longest_word):
        longest_word = line
file.close()
print(longest_word)
```

```python
d = {}
file = open('words_alpha.txt')
for line in file:
    word = line.rstrip()
    word_len = len(word)
    if word_len not in d:
        d[word_len] = 1
    else:
        d[word_len] += 1
file.close()
```

```python
print(d)
```

```python
for word_length in sorted(d):
    print(f'{word_length}\t{d[word_length]}')
```

```python
file = open('Escherichia_coli.tsv')
for line in file:
    if line.startswith('#'):
        continue
    columns = line.split()
    gene_name = columns[0]
    gene_length = int(columns[2]) - int(columns[1]) + 1
    print(f'{gene_name}\t{gene_length}')
file.close()
```
